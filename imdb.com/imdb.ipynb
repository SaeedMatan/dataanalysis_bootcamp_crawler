```python
import pandas as pd
```

```python
data=pd.read_csv("C:\\Users\\E City\\Desktop\\dataa.csv")
df=pd.DataFrame(data)
df.head(15)
```

```python
import re
a=[]
for i in df['gross']:
    t=re.sub('\$','',i)
    t=re.sub('M','',t)
    t=re.sub(',','.',t)
    a.append(float(t))
df['gross']=a
```

```python
b=[]

for i in df['year']:
    t=re.sub('-','',i)
    b.append(t)   
df['year']=b
df
```

```python
ff=df["director"].isin(["Quentin Tarantino" ,"Martin Scorsese" , "Sidney Lumet" 
                        , "David Fincher" ,"Billy Wilder","Alfred Hitchcock"])
df[ff].groupby('director')['rate'].mean().sort_values(ascending=False)
```
